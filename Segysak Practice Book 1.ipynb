```python
import segysak as sg
%matplotlib inline
```

# Create an empty 3D seismic cube

```python
help(sg.create3d_dataset)
```

```python
cube = sg.create3d_dataset(
    dims = (10,15,1000),
    sample_rate = 4,
    vert_domain = 'TWT',
    vert_units  = 'ms'
)
cube
```

# Selection based on Dimensions

```python
select_iline= cube.sel(iline = 10)
select_iline
select_xline= cube.sel(xline = 10)
select_xline
cropped_cube = cube.sel(
    iline = slice(2,9,2),
    xline = slice(4,11,3),
    twt   = slice(700)
)
cropped_cube
cropped_cube = cube.sel(
    iline = cube.iline[:9],
    xline = cube.xline[:11],
    twt   = cube.twt[400:700]
)
cropped_cube
cropped_cube = cube.sel(
    iline = cube.iline[:9],
    xline = cube.xline[:11],
    twt   = 402,
    method = 'nearest'
)
cropped_cube
```
